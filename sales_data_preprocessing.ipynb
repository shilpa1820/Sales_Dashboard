```python
import pandas as pd
```

```python
jan= pd.read_csv("/content/Sales_January_2019.csv")
feb=pd.read_csv("/content/Sales_February_2019.csv")
march= pd.read_csv("/content/Sales_March_2019.csv")
```

```python
jan.head()
```

```python
sales = pd.concat([jan,feb,march], ignore_index=True, axis=0)
```

```python
sales.head()
```

```python
sales.head()
```

```python
sales.isna().sum()
```

```python
sales= sales.dropna()
```

```python
sales.isna().sum()
```

```python
split_1=sales['Purchase Address'].str.split(",",expand=True)
split_1.head()
```

```python
split_2=split_1[2].str.split(" ",expand=True)
split_2.head()
```

```python
sales['postal_code']=split_2[2]
sales.head()
```

```python
sales["postal_code"].unique()
```

```python
sales.isna().sum()
```

```python
sales= sales.dropna()
```
